### 1. Загрузить, посмотреть, определить количество строк и объединить 3 датасета: `marketing_campaign.csv`, `users.csv` и `subscribers.csv`.

```python
df_1 = pd.read_csv("./content/HW_1/marketing_campaign.csv", sep=',')
df_1.head()
```

```python
print(f"Количество строк в датасете df_1: {df_1.shape[0]}")
```

```python
df_2 = pd.read_csv("./content/HW_1/users.csv", sep=',')
df_2.head()
```

```python
print(f"Количество строк в датасете df_2: {df_2.shape[0]}")
```

```python
df_3 = pd.read_csv("./content/HW_1/subscribers.csv", sep=',')
df_3.head()
```

```python
print(f"Количество строк в датасете df_3: {df_3.shape[0]}")
```

```python
# t = pd.merge(df_1, df_2, on="user_id")
df = pd.merge(
    pd.merge(df_1, df_2, on="user_id", how="outer")
    , df_3
    , on="user_id"
    , how="outer"
)

df
```

### 3. Определить эффективность маркетинговых каналов по привлечению платящих игроков.

```python
df.groupby("marketing_channel")["is_retained"].mean().sort_values(ascending=False)
```

Наиболее эффективным маркетинговым каналом по критерию привлечения платящих игроков является `Email`
